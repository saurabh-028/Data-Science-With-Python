# Data Visualization using Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.linspace(0,20,10)
t
```

```python
y=t**2
```

```python
plt.grid(True)
#plt.plot(t,y,color='r',linestyle="--",linewidth=2.5,alpha=0.9,label="wave")   #alpha- transperancy
plt.plot(t,y,color="r",marker="o",markersize=10,label="dot line",markerfacecolor="g")
# plt.xlim(0,30)
# plt.ylim(0,500)
plt.title("Squaring",size=20,color="r")
plt.xlabel("time",size=20)
plt.ylabel("amp")
plt.legend(loc=1)
plt.show()
```

```python
# multiple plots in a single plot
# plt.plot(t,np.sin(t),color="g",label="sin wave")
# plt.show()
# plt.plot(t,np.cos(t),color="r",label="cos wave")
# plt.show()
```

```python
# multiple plot using subplot

plt.subplot(2,1,1)
plt.plot(t,np.sin(t))
plt.subplot(2,1,2)
plt.plot(t,np.cos(t))
```

```python
# scatterr plot

#compare the continuous/numerical values

plt.scatter(t,y,marker="X",color="g")
plt.show()

```

```python
# Bar plot
#between numerical and categorial(object)
x=["Apple","orange","banana","grapes"]
y=[70,85,25,30]
plt.barh(x,y,color="r")
```

```python
# histogram
data=np.random.randn(1000)
plt.hist(data)
```

```python
# pie chart

sales=[10,20,8,35,30,25]
agent=["Raj","john","kumar","sugan","dev","jagan"]  # Raj=10/  (10+20+8+35+30+25)
#explode1=[0.5,0,0,0,0.2,0]
plt.pie(sales,labels=agent,explode=[0.5,0,0,0,0.2,0],autopct="%.1f%%")
plt.show()
```

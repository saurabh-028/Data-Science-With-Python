# Pandas merge,concat,duplicate

```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({"Name":["john","ram","jack","dan","jeff","kumar"],"ID":[10,20,30,25,28,32]})
```

```python
df2=pd.DataFrame({"student":["gaya","ram","jack","kumar",],"ID":[40,20,30,25]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on="ID")
```

```python
pd.merge(df1,df2,how="left",on="student")
```

```python
pd.merge(df1,df2,how="right",on="student")
```

```python
pd.merge(df1,df2,how="outer",on="ID")
```

```python
# conacat
pd.concat([df1,df2],axis=1)
```
